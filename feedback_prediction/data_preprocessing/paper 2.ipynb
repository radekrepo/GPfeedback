```python
import requests
import os
import sys


"""
This code was used with Python 3.6.3


scraping data from a larger number of subpages

NOTE: The API key "JROLO" is not a genuine NHS Choices API key. 
To use this code, please apply to NHS Choices for your own API key. Please
also take into account that this code may not always conform to fair use
policy of NHS Choices (I've had problems with opening too many pages per hour)

I used 'C:\\Users\\user\\Desktop' as my initial current working directory
(double forward slashes are needed to make the code run correctly)

The start date for scraping of articles (the earlier date considered for scraping) can be chosen flexibly.
At the time of writing, NHS Choices (data provider) only makes available the last 2 years of comments.

"""


strony = range(1443,3450) #range of pages with NHS GP reviews


temp_folder = "GP_comments_2017" # this code is for scraping articles from 2017
temppath = "C:\\Users\\user\\Desktop" + "\\" + temp_folder

if os.path.isdir(temppath) == False:
    os.makedirs(temppath)

if os.getcwd() != temppath:
    os.chdir(temppath)

for strona in strony:
    adres = "http://v1.syndication.nhschoices.nhs.uk/organisations/gppractices/commentssince/2017/1/1?apikey=JROLO&page="
    nazwa_pliku = "gp2017_comments"
    adres = adres + str(strona)
    nazwa_pliku = nazwa_pliku + str(strona) + ".txt"
    r = requests.get(adres)
    with open(nazwa_pliku, "wb") as f:
        f.write(str.encode(r.text))
    print ("downloaded " + str(strona) + " page of comments")

print ("ok")
```

```python
"""

transforming .txt files containing xml-formatted data into spreadsheet in .csv format


question indices (questions 8-13 relate to GP practices, in .csv file they're named questions 9-14):
0 - 'How likely are you to recommend this pharmacy to friends and family if they needed similar care or treatment?'
1 - 'How likely are you to recommend this optician to friends and family if they needed similar care or treatment?'
2 - 'How satisfied are you with the cleanliness of the area you were treated in?'
3 - 'How satisfied are you that the hospital staff worked well together?'
4 - 'How satisfied are you that you were treated with dignity and respect by the staff?'
5 - 'How satisfied are you that you were involved in decisions about your care?'
6 - 'How satisfied are you that the hospital provides same-sex accommodation?'
7 - 'How likely are you to recommend this hospital to friends and family if they needed similar care or treatment?'
8 - 'Are you able to get through to the surgery by telephone?'
9 - 'Are you able to get an appointment when you want one?'
10 - 'Do the staff treat you with dignity and respect?'
11 - 'Does the surgery involve you in decisions about your care and treatment?'
12 - 'How likely are you to recommend this GP surgery to friends and family if they needed similar care or treatment?'
13 - 'This GP practice provides accurate and up to date information on services and opening hours'
14 - 'How satisfied are you with the time you have to wait for an appointment?'
15 - 'How satisfied are you that you are you treated with dignity and respect by staff?'
16 - 'How satisfied are you that the dental surgery involves you in decisions about your treatment?'
17 - 'How satisfied are you with the information given by the surgery on the cost of your NHS treatment?'
18 - 'How satisfied are you with the outcome of your treatment?'
19 - 'How likely are you to recommend this dentist to friends and family if they needed similar care or treatment?'
20 - 'The environment where I was treated was?'
21 - 'The Clinic  staff worked well together?'
22 - 'I was treated with dignity and respect by the Clinic  staff?'
23 - 'I was involved with decisions about my care?'
24 - 'Friends and Family'
25 - 'How likely are you to recommend this care provider to friends and family if they needed similar care or treatment?'
26 - 'NHS Clinic s must provide same-sex accommodation. How satisfied were you that this Clinic  did so?'
27 - 'Overall recommendation '

Other questions are relevant when analysing reviews of other types of NHS services, 
such as hospitals, care centers and dentists. Not all types of NHS insitutions received end user feedback.
For more information, check out http://v1.syndication.nhschoices.nhs.uk/organisations?apikey=JROLO

Profile pages for individual GP practices, where ODS codes and other data are available, can be retrieved as well
For example, GP practice with NHS ID 17657 can be retrieved from
http://v1.syndication.nhschoices.nhs.uk/organisations/gppractices/17657?apikey=JROLO

"""

import os
import re
import glob
import csv


output_csv = "comments_2017.csv"

miesiace = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]




foldery = ["GP_comments_2017"] # folders with comments according to NHS institution type
# wiersz_zero = ["MyDatabaseID", "category", "NHS_review_ID", "type", "commenter"]
pytania =['How likely are you to recommend this pharmacy to friends and family if they needed similar care or treatment?', 
          'How likely are you to recommend this optician to friends and family if they needed similar care or treatment?', 
          'How satisfied are you with the cleanliness of the area you were treated in?', 
          'How satisfied are you that the hospital staff worked well together?', 
          'How satisfied are you that you were treated with dignity and respect by the staff?', 
          'How satisfied are you that you were involved in decisions about your care?', 
          'How satisfied are you that the hospital provides same-sex accommodation?', 
          'How likely are you to recommend this hospital to friends and family if they needed similar care or treatment?', 
          'Are you able to get through to the surgery by telephone?', 'Are you able to get an appointment when you want one?', 
          'Do the staff treat you with dignity and respect?', 
          'Does the surgery involve you in decisions about your care and treatment?', 
          'How likely are you to recommend this GP surgery to friends and family if they needed similar care or treatment?',
          'This GP practice provides accurate and up to date information on services and opening hours', 
          'How satisfied are you with the time you have to wait for an appointment?', 
          'How satisfied are you that you are you treated with dignity and respect by staff?', 
          'How satisfied are you that the dental surgery involves you in decisions about your treatment?', 
          'How satisfied are you with the information given by the surgery on the cost of your NHS treatment?', 
          'How satisfied are you with the outcome of your treatment?', 
          'How likely are you to recommend this dentist to friends and family if they needed similar care or treatment?', 
          'The environment where I was treated was?', 'The Clinic  staff worked well together?', 
          'I was treated with dignity and respect by the Clinic  staff?', 'I was involved with decisions about my care?', 
          'Friends and Family', 
          'How likely are you to recommend this care provider to friends and family if they needed similar care or treatment?',
          'NHS Clinic s must provide same-sex accommodation. How satisfied were you that this Clinic  did so?', 
          'Overall recommendation']
tymcz = (len(pytania) + 1)*3 
list_pyt = range(tymcz)
list_pyt = [ "" for a in list_pyt]
rodzaje_uslug = []
partia = 0
starter = 1 #number generator for comment IDs
for i in foldery:
    i = 'c:\\users\\user\\desktop' + '\\' + i
    os.chdir(i)
    if os.path.isfile(output_csv) == True:
        os.remove(output_csv)
    licznik = 0
    wiersze = []
    for indeks1, filer in enumerate(glob.glob("*.txt")):
        licznik = licznik + 1
        if indeks1 < 20000:
            with open(filer, "r", encoding="utf8") as f:
                okej = f.read()
                if '<entry>' in okej:
                    start_lista = [a.start() for a in re.finditer('<entry>', okej)]
                    koniec_lista = [a.end() for a in re.finditer('</entry>', okej)]
                else:
                    start_lista = [a.start() for a in re.finditer('<item>', okej)]
                    koniec_lista = [a.end() for a in re.finditer('</item>', okej)]
                komentarze = []
            # list positions of segments in downloaded .xml files which correspond to individual comments
                for i, value in enumerate(start_lista):
                    komentarz = []
                    komentarz.append(start_lista[i])
                    komentarz.append(koniec_lista[i])
                    komentarze.append(komentarz)
            #   CREATE A LIST CONTAINING INFORMATION ABOUT EACH COMMENT
                for indeks, punkty_start_koniec in enumerate(komentarze):
                    tresc = okej[punkty_start_koniec[0]:punkty_start_koniec[1]]
            #     first element - comment ID in the dataset
                    wiersz = []
                    wiersz.append(filer)
                    id = str(starter) + str(".2017") #comment ID in the dataset being created
                    wiersz.append(id)
                    starter = starter + 1
            #     second element - comment category (e.g. about dentists, pharmacists)
                    a = "http://v1.syndication.nhschoices.nhs.uk/organisations/"
                    b = "/commentssince"
                    if a and b in okej:
                        jeden = okej.index(a)
                        dwa = okej.index(b)
                        digit = jeden + len(a)
                        wynik = okej[digit:dwa]
                        wiersz.append(wynik)
            # third element - comment ID in NHS database
                    if '<id>' in tresc:
                        a = tresc.find('<id>') + len('<id>')
                        b = tresc.find("</id>")
                        fin = str(tresc[a:b])
                        znak = fin.find("R" or "r")
                        if znak == -1:
                            wiersz.append(fin)
                        else:
                            fin = fin[:-1]
                            wiersz.append(fin)
                    else:
                        a = tresc.find('<guid isPermaLink="false">') + len('<guid isPermaLink="false">')
                        b = tresc.find("</guid>")
                        fin = str(tresc[a:b])
                        znak = fin.find("R" or "r")
                        if znak == -1:
                            wiersz.append(fin)
                        else:
                            fin = fin[:-1]
                            wiersz.append(fin)
            # fourth element - type of comment (comment, reply or deletion)
                    if '<id>' in tresc:
                        a = tresc.find('<category term="') + len('<category term="')
                        b = tresc.find('" label="Whether this is a comment, reply or deletion" scheme="commentType" />')
                        wiersz.append(tresc[a:b])
                    else:
                        a = tresc.find('<category domain="commentType">') + len('<category domain="commentType">')
                        b = tresc.find('</category>')
                        wiersz.append(tresc[a:b])
            # fifth element - commented-on institution's ID in NHS database
                    if '<id>' in tresc:
                        a = tresc.find('"Organisation commented on" length="1000" href="http://v1.syndication.nhschoices.nhs.uk/organisations/') + len('"Organisation commented on" length="1000" href="http://v1.syndication.nhschoices.nhs.uk/organisations/')
                        tymcz = tresc[a:]
                        a = tymcz.find("/") + len("/")
                        b = tymcz.find("?apikey=JROLO")
                        wiersz.append(tymcz[a:b])
                    else:
                        a = tresc.find('"Organisation commented on" length="1000" href="http://v1.syndication.nhschoices.nhs.uk/organisations/gppractices/') + len('"Organisation commented on" length="1000" href="http://v1.syndication.nhschoices.nhs.uk/organisations/gppractices/')
                        b = tresc.find("?apikey=JROLO")
                        wiersz.append(tresc[a:b])
            # sixth element - comment author
                    if '<id>' in tresc:
                        a = tresc.find('<author>')
                        b = tresc.find("</author>")
                        tymcz = tresc[a:b]
                        a = tymcz.find("<name>") + len("<name>")
                        b = tymcz.find("</name>")
                        if tymcz[a:b] != "Anonymous":
                            wiersz.append(tymcz[a:b])
                        else:
                            wiersz.append("nan")
                    else:
                        a = tresc.find('<a10:author>')
                        b = tresc.find("</a10:author>")
                        tymcz = tresc[a:b]
                        a = tymcz.find("<a10:name>") + len("<a10:name>")
                        b = tymcz.find("</a10:name>")
                        if tymcz[a:b] != "Anonymous":
                            wiersz.append(tymcz[a:b])
                        else:
                            wiersz.append("nan")
            # seventh element - comment title
                    if '<id>' in tresc:
                        a = tresc.find('<title type="text">') + len('<title type="text">')
                        b = tresc.find("</title>")
                        fin = str(tresc[a:b])
                        if fin != 'Rating without comment' and wiersz[4] != 'deletion':
                            wiersz.append(fin)
                        else:
                        # add also 8th element if no comment posted or if it's about comment deletion
                            wiersz.extend(['nan', 'nan'])
                    else:
                        a = tresc.find('<title>') + len('<title>')
                        b = tresc.find("</title>")
                        fin = str(tresc[a:b])
                        if fin != 'Rating without comment' and wiersz[4] != 'deletion':
                            wiersz.append(fin)
                        else:
                        # add also 8th element if no comment posted or if it's about comment deletion
                            wiersz.extend(['nan', 'nan'])
            # 8th element - comment text
                    if len(wiersz) == 8:
                        if '<id>' in tresc:
                            a = tresc.find('[CDATA[') + len('[CDATA[')
                            b = tresc.find("]]>")
                            tymcz = tresc[a:b].replace('\n',' ')
                            tymcz = tymcz.replace('&apos;','')
                            tymcz = tymcz.replace('','')
                            tymcz = tymcz.replace('&amp;quot;','')
                            tymcz = tymcz.replace("&amp;amp;",' and ')
                            wiersz.append(tymcz)
                        else:
                            a = tresc.find('[CDATA[') + len('[CDATA[')
                            b = tresc.find("]]&gt")
                            tymcz = tresc[a:b].replace('\n',' ')
                            tymcz = tymcz.replace('&amp;apos;','')
                            tymcz = tymcz.replace('','')
                            tymcz = tymcz.replace('&amp;quot;','')
                            tymcz = tymcz.replace("&amp;amp;",' and ')
                            wiersz.append(tymcz)
            # elements 9, 10, 11 - date (day, month, year)
                    if '<id>' in tresc:
                        a = tresc.find('<published>') + len('<published>')
                        b = tresc.find("</published>")
                        tymcz = tresc[a:b]
                        # day of the month
                        wiersz.append(tymcz[8:10])
                        # month
                        wiersz.append(tymcz[5:7])
                        # year
                        wiersz.append(tymcz[0:4])
                    else:
                        a = tresc.find('<pubDate>') + len('<pubDate>')
                        b = tresc.find("</pubDate>")
                        tymcz = tresc[a:b]
                        # day of the month
                        wiersz.append(tymcz[5:7])
                        # month
                        msc = tymcz[8:11]
                        msc = miesiace.index(msc) + 1
                        wiersz.append(str(msc))
                        # year
                        wiersz.append(tymcz[12:16])
            # collecting numeric Likert-scale feedback
                    pocz = [a.end() for a in re.finditer('<rating>', tresc)]
                    kon = [a.start() for a in re.finditer('</rating>', tresc)]
                    list_pyt2 = list_pyt[:]
                    for ind, wart in enumerate(pocz):
                        pytanie = tresc[(wart):kon[ind]]
                        a = pytanie.find('<questionText>') + len('<questionText>')
                        b = pytanie.find("</questionText>")
                        tymcz = pytanie[a:b]
                        for i, row in enumerate(pytania):
                            if tymcz != row:
                                pass
                            else:
                                indeks = i*3
                                list_pyt2[indeks] = indeks/3 + 1
                                a = pytanie.find('<answerText>') + len('<answerText>')
                                b = pytanie.find("</answerText>")
                                list_pyt2[(indeks+1)] = pytanie[a:b]
                                a = pytanie.find('<answerMetric value="') + len('<answerMetric value="')
                                odpowiedz = pytanie[a:(a+4)]
                                b = odpowiedz.find('"')
                                list_pyt2[(indeks+2)] = odpowiedz[:b]
                    wiersz.extend(list_pyt2)
                    wiersze.append(wiersz)
                if licznik < 100 and filer != glob.glob("*.txt")[-1]:
                    pass
#                     print('indeks1 ' + str(indeks1))
#                     wiersze.append(wiersz)
                else:
#                     print('partia')
                    partia = partia + 1
                    print("writing to .csv batch " + str(partia) + " (100 pages of reviews per batch)")
                    with open(output_csv,'a', newline='', encoding='utf-8') as final:
                        writer = csv.writer(final)
                        for x in wiersze:
                            writer.writerow(x)
                    licznik = 0
                    wiersze = []
#                     wiersze.append(wiersz)


    os.chdir('c:\\users\\user\\desktop')


print ("done")
```

```python
"""

Merging new reviews with old reviews dataset. 

Editions of the new dataset prior to merging
- delete column with comment titles
- delete column with file names
- reorganize questions (retain only Likert-scale responses)
- add row with column titles

Merging activities:
- combine scraped data from 2017 with older dataset
"""

import os
import csv


output_csv = "all_gp_comments.csv"

if os.path.isfile(output_csv) == True:
    os.remove(output_csv)

with open(output_csv,'a', newline='', encoding='utf-8') as final:
    writer = csv.writer(final)
    with open('old_gp_comments.csv','r', newline='', encoding='utf-8') as doc:
        reader = csv.reader(doc)
        for i, x in enumerate(reader):
            if len(x[0]) != 0:
                writer.writerow(x)
    with open('GP_comments_2017\\comments_2017.csv','r', newline='', encoding='utf-8') as doc2:
        reader2 = csv.reader(doc2)
        for i, x in enumerate(reader2):
            if x[4] == 'comment':
                x2 = []
                x2.append(x[1])
                x2.append(x[5])
                x2.extend(x[8:12])
                tm = [x[38],x[41],x[44],x[47],x[50],x[53]]
                tm2 = []
                for xx in tm:
                    if len(xx) > 0:
                        tm2.append(xx)
                    else:
                        tm2.append("nan")
                x2.extend(tm2)
                writer.writerow(x2)



print('finished')
```

```python
"""


- remove:
    - duplicate reviews (which accumulate due to real-time updates to NHS database while scraping)
    - reviews with no comment text 
    - reviews with no Likert-scale responses


"""

import os
import csv
import pandas as pd
import datetime

output_csv = "all_gp_comments4.csv"

if os.path.isfile(output_csv) == True:
    os.remove(output_csv)

print ("removing reviews with missing data")
bd = 0
with open('all_gp_comments.csv','r', newline='', encoding='utf-8') as doc:
    reader = csv.reader(doc)
    with open("all_gp_comments2.csv",'a', newline='', encoding='utf-8') as docout:
        writer = csv.writer(docout)
        for i, x in enumerate(reader):
            if i != 0:
                if x[2] != 'nan':
                    if len(list(set(x[6:]))) != 1 or list(set(x[6:]))[0] != 'nan':
                        a = datetime.date(int(x[5]),int(x[4]),int(x[3]))
                        a = datetime.date.toordinal(a) # a becomes date represented as a single ordinal number
                        x.append(a)
                        writer.writerow(x)
                    else:
                        bd = bd + 1 #count reviews excluded due to lack of Likert-scale ratings
                else:
                    bd = bd + 1 #count reviews excluded due to missing review text
            elif i == 0:
                x.append("datetime")
                writer.writerow(x)


print("number of reviews removed due to missing data:")
print(bd)
print()

# sort reviews by date represented as single ordinal numbers
df = pd.read_csv("all_gp_comments2.csv", sep=',')
df = df.sort_values(['datetime'], ascending=[True])
df.to_csv('all_gp_comments3.csv', sep=',')



print('removing duplicate reviews')
du = 0
# Duplicates: same date of posting, same review text, same commented-on GP practice ID
with open('all_gp_comments3.csv','r', newline='', encoding='utf-8') as doc:
    reader = csv.reader(doc)
    with open(output_csv,'a', newline='', encoding='utf-8') as final:
        writer = csv.writer(final)
        data = 0
        sl = {}
        for i, x in enumerate(reader):
            if i > 0:
                if x[-1] == data:
                    if x[2] in sl.keys():
                        if x[3] != sl[x[2]]:
                            sl[x[2]] = x[3]
                            writer.writerow(x[1:])
                        else:
                            du = du + 1
                    else:
                        sl[x[2]] = x[3]
                        writer.writerow(x[1:])
                else:
                    writer.writerow(x[1:])
                    data = x[-1]
                    sl = {}
                    sl[x[2]] = x[3]
            elif i == 0:
                writer.writerow(x[1:])

print("number of duplicates removed:")
print(du)

# deletion of intermediate .csv spreadsheets
os.remove('all_gp_comments2.csv')
os.remove('all_gp_comments3.csv')


print("done")
```
