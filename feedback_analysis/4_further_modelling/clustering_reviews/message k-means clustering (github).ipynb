```python
"""

This is legacy code written for Python 2.7


K-means clustering

Import:
    "topics_kmeans.csv" - 145388 comments represented through thematic proportion 
                            within messages. The input is the 'regdata' object produced
                            in R in the process on computing LDA topic model results.
    
Export:
    "cluster_centres.csv" - topic structures of clusters of messages
    "comment_cluster_labels.csv" - which comment belongs to which cluster

"""

import sklearn.cluster
from numpy import genfromtxt
import warnings
import csv
import os

# os.chdir('c:\\users\\user\\desktop\\')
# os.getcwd()

wchod = 'topics_kmeans.csv'
klastry = "cluster_centres.csv" # "centra_klastrow, n=100.csv"
etykiety = "comment_cluster_labels.csv"

if os.path.isfile(klastry):
    os.remove(klastry)
if os.path.isfile(etykiety):
    os.remove(etykiety)

my_data = genfromtxt(wchod, delimiter=',')


# model = sklearn.cluster.MiniBatchKMeans() LUB KMeans()
with warnings.catch_warnings(): # Z "with" UNIKAM DRUKOWANIA DO INTERFEJSU DeprecationWarning
    warnings.simplefilter("ignore", category=DeprecationWarning)
    model = sklearn.cluster.KMeans(n_clusters=100,random_state=10) #batch_size=10000, init_size = 10
    print model.get_params
    model.fit(X=my_data)


print "Labels"
# print model.labels_
# print

with open(etykiety, 'ab') as dok:
    writer = csv.writer(dok, delimiter=',')
    for i,v in enumerate(model.labels_):
        writer.writerow([v])

print "cluster centres"
# print model.cluster_centers_
# print

with open(klastry, 'ab') as dok:
    writer = csv.writer(dok, delimiter = ',')
    for i,v in enumerate(model.cluster_centers_):
        l = list(model.labels_)       
        ile_razy = l.count(i) #liczenie ile wiadomości podporządkowano danemu klastrowi
        v = list(v)
        v.append(ile_razy)
        writer.writerow(v)


print "ok"

# print "inercja"
# print model.inertia_


# help(sklearn.cluster.MiniBatchKMeans)
```
